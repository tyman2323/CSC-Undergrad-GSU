```python
import numpy as np
labels = []
def kmeans(x,k,iters):#kmeans implementation to attain y labels 
    centroids = x[np.random.choice(len(x), k, replace=False)]
    for z in range(iters):
        labels = np.argmin(np.linalg.norm(x[:, np.newaxis] - centroids, axis=2), axis=1)
        new_centroids = np.array([x[labels == i].mean(axis=0) for i in range(k)])
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids
    return labels
y = kmeans(datahot.values, 3, 1000)
print(y)
```

```python
from sklearn.model_selection import train_test_split
def fit(x,y):#fit function to train 
    classes, counts = np.unique(y, return_counts=True)
    classprob = counts / len(y)
    featureprob = []

    for c in classes:
        labelisc = (y == c)
        classcount = np.sum(labelisc)
        currfeatureprob = (x[labelisc] + 1) / (classcount + x.shape[1])  # laplace smoothing
        featureprob.append(currfeatureprob)

    return classprob, featureprob, classes

def predict_naivebayes(x, classprobfinal, featureprobfinal): #uses teh testing set for making predictions 
    predictions = []

    for xx in x:
        classprobx = np.log(classprobfinal)
        for i, c in enumerate(classes):
            currfeatureprob = np.log(featureprobfinal[i])
            currprediction = np.sum(currfeatureprob * xx)
            classprobx[i] += currprediction

        prediction = classes[np.argmax(classprobx)]
        predictions.append(prediction)

    return np.array(predictions)#returns all predictions

xtrain, xtest, ytrain, ytest = train_test_split(datahot.values, y, test_size=0.2, random_state=42)#used to get training and testing sets
classprobfinal, featureprobfinal, classes = fit(xtrain, ytrain)
print(classprobfinal,featureprobfinal)
pred = predict_naivebayes(xtest, classprobfinal, featureprobfinal)
print(pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, pred))#used to attain accuracy
```

```python
def pdf(x, mean, std):
    epsilon = .0001#epsilon for divide by 0
    exponent = np.exp(-((x - mean)**2) / (2 * (std**2 + epsilon)))
    f = (1 / (np.sqrt(2 * np.pi) * (std + epsilon))) * exponent
    return np.where(f == 0, epsilon, f)

def gaussianfit(x, y):
    classes, counts = np.unique(y, return_counts=True)
    classprob = counts / len(y)
    featuremeans = []#used for gaussian
    featurestds = []#used for gaussian

    for c in classes:
        labelisc = (y == c)
        data = x[labelisc]
        featuremeans.append(np.mean(data, axis=0))
        featurestds.append(np.std(data, axis=0))
    return classes, classprob, featuremeans, featurestds

def predictgaussian(x, classes, classprobfinal, featuremeansfinal, featurestdsfinal):
    predictions = []

    for xx in x:
        classprobsx = np.log(classprobfinal)
        for i, c in enumerate(classes):
            #if np.any(xx == 0) or np.any(featuremeansfinal[i] == 0) or np.any(featurestdsfinal[i] == 0):
              #  print('Found a 0 at', xx)
            currfeature = np.sum(np.log(pdf(xx, featuremeansfinal[i], featurestdsfinal[i])))
            classprobsx[i] += currfeature

        predictedclass = classes[np.argmax(classprobsx)]
        predictions.append(predictedclass)

    return np.array(predictions)


classes, classprobfinal, featuremeansfinal, featurestdsfinal = gaussianfit(xtrain, ytrain)

print(classprobfinal, '\n-----------------', featuremeansfinal, '\n\n', featurestdsfinal)

pred = predictgaussian(xtest, classes, classprobfinal, featuremeansfinal, featurestdsfinal)
print(pred)
```
