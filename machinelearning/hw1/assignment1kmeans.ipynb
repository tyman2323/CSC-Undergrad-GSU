```python
import pandas as pd
import numpy as np


def distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))


def center(data, k):
    index = np.random.choice(len(data), k, replace=False)
    centers = data[index]
    return centers


def cluster(data, centers):
    distances = np.array([[distance(point, center) for center in centers] for point in data])
    return np.argmin(distances, axis=1)
#attain distances and place points into clusters

def updatecenters(data, label, k):
    centers = np.zeros((k, data.shape[1]))
    for x in range(k):
        points = data[label == x]
        if len(points) > 0:
            centers[x] = np.mean(points, axis=0)
    return centers
#provide new centers to update the cluster 

def score(data, labels):
    distances = np.array([[distance(data[i], data[j]) for j in range(len(data))] for i in range(len(data))])
    a = np.array([np.mean(distances[i, labels == labels[i]]) for i in range(len(data))])
    b = np.array([np.min([np.mean(distances[i, labels == j]) for j in range(len(set(labels))) if j != labels[i]]) for i in range(len(data))])
    s = (b - a) / np.maximum(a, b)
    return np.mean(s)


dataarr = df[['3rd', '4th']].to_numpy()
kval = [5]#testing to see if it can work with one then upscale to more ks

for k in kval:
    centerr = center(dataarr, k)
    pastlabel = np.zeros(len(dataarr))
    while True:#loops continuously untill there is no more change
        labels = cluster(dataarr, centerr)
        if np.array_equal(labels, pastlabel):
            print(k)
            scores = score(dataarr, labels)  
            print("Silhouette Score:", scores)
            break
        else:
            centerr = updatecenters(dataarr, labels, k)
            pastlabel = labels
scores = score(dataarr,labels)

```
