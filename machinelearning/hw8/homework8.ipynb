```python
import numpy as np

def decision(w, b, x):
    return np.dot(x, w) + b

def hingeloss(w, b, x, y):
    margin = 1 - y * decision(w, b, x)
    return np.maximum(0, margin)

def objective(w, b, C, x, y):
    return 0.5 * np.dot(w, w) + C * np.mean(hingeloss(w, b, x, y))

def gradients(w, b, C, x, y):
    margin = 1 - y * decision(w, b, x)
    indicator = (margin > 0).astype(int)
    dw = w - C * np.dot(x.T, y * indicator) / len(y)
    db = -C * np.sum(y * indicator) / len(y)
    return dw, db

def trainsvc(x, y, C, learning_rate, num_epochs):
    w = np.zeros(x.shape[1]) 
    b = 0 
    for epoch in range(num_epochs):

        dw, db = gradients(w, b, C, x, y)
        w -= learning_rate * dw
        b -= learning_rate * db

        loss = objective(w, b, C, x, y)
        print(f'Epoch {epoch}: Loss = {loss}')
    print("Training completed!")
    return w, b



wsvc, bsvc = trainsvc(x, y, 1.0, .01, 1000)




```
